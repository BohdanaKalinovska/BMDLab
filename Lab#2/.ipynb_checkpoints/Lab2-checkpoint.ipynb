Завдання:

1. Створити вектор розміром 11 зі значенням від 0 до 5. Розрахувати на його основі вектор y(y=x^2)

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

2. Побудувати графіки

2.1 Функціональний метод

```python
plt.plot(x, y) #build plot
plt.xlabel('X Label') #add name of x axis
plt.ylabel('Y Label') #add name of y axis
plt.title('Title') #add title
```

```python
#Для побудови декількох графіків поруч використовують команду subplot 
plt.subplot(1, 2, 1) #create first plot: 1 row, 2 colums, 1st plot
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2) #create second plot: 1 row, 2 colums, 2d plot
plt.plot(y, x, 'b')
```

2.2 Об'єктно-орієнтований метод

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1]) #(від 0 по осі x, від 0 по осі y, ширина поля графіку, висота поля графіку )

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

 Одночасна побудова декількох графіків, додаємо їх до об'єкта з вказанням області відображення:

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')

axes2.plot(x, y, 'b')
axes2.set_xlabel('Y label')
axes2.set_ylabel('X label')
axes2.set_title('Title')
```

Для використання функції subplot при створенні графіка як об'єкта, його ініціалізація має бути виконана наступним чином:

```python
#Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

Вказуємо кількість рядків і стовпців під час створення об'єкта subplots():

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2 )

axes[0].plot(x,y)
axes[1].plot(y,x)
```

Випадок часткового накладання графіків:

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4 )
```

```python
#Для того щоб уникнути накладання варто використовувати команду plt.tight_layout()
fig, axes = plt.subplots(nrows = 4, ncols = 4 )
plt.tight_layout()
```

Додавання легенди до графіку з використанням команди .legend()

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

Змінення розміру графіка та DPI:

```python
fig = plt.figure(figsize=(8,4), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

При створенні графіку можна задати наступні характеристики:

- колір графіку 

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, color = 'b') #blue line
ax.plot(x, x**3, color = 'g') #green line
```

- тип та ширина лінії

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 2, ls = '--')
```

- тип, розмір та колір маркерів

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, marker = 'o', markersize = 10, markerfacecolor = 'y')
```

Для збереження побудованого графіку в форматах PNG, JPG, EPS, SVG, PGF, PDF використовують команду savefig:

```python
fig.savefig("filename.png")
```

Існує велика кількість спеціальних графіків, які можна побудувати в Matplotlib:

- точкова діаграма

```python
plt.scatter(x,y)
```

- гістограма

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

- box-plot діаграма

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data, vert = True, patch_artist = True);
```
