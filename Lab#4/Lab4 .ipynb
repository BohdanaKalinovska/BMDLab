1. Сгенеруйте набір даних, що містить не менше 6 змінних. Частину змінних згенеруйте за нормальний розподілом, частину - за будь-яким іншим розподілом. Кількість спостережень - 20

```python
#Імпорт бібліотек

import pandas as pd 
import numpy as np 
import scipy.stats as sts 
```

```python
#Нормальний(Гаусовський): random.normal(loc,scale,size) 
                    #loc- Середнє значення ("центр") розподілу,
                    #scale - Стандартне відхилення ("ширина") розподілу,
                    #size - Кількість значень
#Рівномірний: random.normal(low,high,size) 
                    #low - Нижня межа вихідного інтервалу. Усі згенеровані значення >=low (За замовчуванням 0)(необов'язковий),
                    #high - Верхня межа інтервалу виводу. Усі створені значення будуть <=hight (За замовчуванням 1)(необов'язковий),
                    #size - Кількість значень
#Біономінальний: random.binomial(n, p, size)
                    #n - Кількість випробувань n>=0,
                    #p - Ймовірність успіху p>=0 and p<=1,
                    #size - Кількість значень
#Стандартний нормальний: random.standard_normal(size)
                    #size - Кількість значень

data = pd.DataFrame({
    'Var_1(normal)': np.random.normal(loc=10, scale=5, size=20),
    'Var_2(normal)': np.random.normal(loc=5, scale=10, size=20),
    'Var_3(uniform)': np.random.uniform(size=20),
    'Var_4(uniform)': np.random.uniform(size=20),
    'Var_5(binom)': np.random.binomial(n=50, p=0.9, size=20),
    'Var_6(st_normal)': np.random.standard_normal(size=20)
})

data
```

2. Створіть функцію, що реалізує критерій Шапіро-Уілка та застосуйте її до кожної змінної набору данних

```python
data = pd.DataFrame({key: sorted(values) for key, values in data.transpose().iterrows()})
```

```python
#За формулою шукаємо суму x' = 1/n * sum(x')    - сума кожного стовпця
def Value (arr,n):
    x_sum =0
    for i in range(n):
        x_sum +=arr[i]
    return x_sum/n
F_value_1 = {i: Value(data[i], 20) for i in data}
```

```python
#Шукаємо S^2 за формулою S^2 = sum((x - x')^2 - від кожного значення віднімаємо суму того ж стопця, беремо те що вийшло в квадрат і сумуємо всі значення
def Sq(arr, n, x):
    S2 = 0
    for i in range(n):
        S2 += ((arr[i] - x)**2)
    return S2
S_s2 = {i: Sq(data[i], 20, F_value_1[i]) for i in data}
```

```python
#Дані з таблиці
a_i = [0.4734, 0.3211, 0.2565, 0.2085, 0.1686, 0.1334, 0.1013, 0.0711, 0.0422, 0.0140] 
#За формулою W = 1/S^2 * sum(a_i) * (x_n_i - x_i)
def W (arr, n, S_s2):
    tmp = 0
    for i in range(n//2):
        tmp +=(a_i[i] * (arr[(n-1)-i] - arr[i]))
    return tmp**2 /S_s2

W_val = {i: W(data[i], 20, S_s2[i]) for i in data}
W_f = {i: sts.shapiro(data[i]) for i in data}

print("Практичне значення: W_val = ", W_val)
print(' ')
print("За вбудованою формулою: W_f = ", W_f)
```

3. Сгенеруйте пусту таблицю, в якій кількість рядків - це кількість змінних + 1. Кількість стовбчиків - 3. Назва стовпчиків - Variable, Statistics, Distribution

```python
headers = ['Variable', 'Statistics', 'Distribution']
new_table_data = pd.DataFrame(columns = headers)
new_table_data
```

4. Занести в стовпчик Variable назви змінних

```python
new_table_data["Variable"] = data.keys()
new_table_data
```

5. Занести в стовпчик Distribution результати роботи криетрію Шапіро-Уілка. Для змінних, які розподілені за нормальним законом вносити "normal", в протилежному вигляді - "non-normal"

```python
new_table_data["Distribution"] = [('normal ' if 'normal' in dist else 'non-normal ')
                          for dist in new_table_data["Variable"]]
new_table_data
```

6. Розрахувати для всіх змінних, що розподілені за нормальним законом розрахувати середне значення та стандартне відхилення. Записати в таблицю. Форма запису: Середнє [стандартне відхилення]

```python
for i in range (new_table_data.shape[0]):
    norm = new_table_data["Variable"][i]
    if 'normal' in norm:
        new_table_data["Statistics"][i] = f".{data[norm].mean()} [ {data[norm].std()} ]"
new_table_data
```

7. Розрахувати для всіх змінних, що розподілені за розподілом відсутнім від нормального розрахувати медіану та інтерквартильний розмах. Записати в таблицю. Форма запису: Медіана [25%; 75%]

```python
for i in range (new_table_data.shape[0]):
    non_normal = new_table_data["Variable"][i]
    if 'normal'not in non_normal:
        new_table_data["Statistics"][i] = f".{data[non_normal].median()} \
        [{data[non_normal].quantile(0.25)}; {data[non_normal].quantile(0.75)}]"

new_table_data
```
